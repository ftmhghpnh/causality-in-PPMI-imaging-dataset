```python
csv_info_pd = pd.read_csv('/w/246/gzk/PPMI/codes/first_model_csv_info.csv')
len(csv_info_pd)
# csv_info_pd.head()
```

```python
series_identifier_lists = csv_info_pd['seriesIdentifier'].tolist()
len(series_identifier_lists)
```

```python
base_path_prob_map = '/w/247/gzk/MRI_PD_sieved_task5/'
```

```python
images_list = os.listdir(base_path_prob_map)
images_list[0].split('_')[-2][1:]
```

```python
series_image_path_dic ={}
for unique_image in series_identifier_lists:
    series_image_path_dic[unique_image] = []
    for image in images_list:
        if str(unique_image) in image:
            if ('c1' in image) or ('c2' in image) or ('c3' in image):
#             if 'c2' in image:
                series_image_path_dic[unique_image].append(image)
    
```

```python
series_image_path_dic
```

```python
prob_maps = []
prob_maps_identifiers = []

for unique_image in series_identifier_lists:
#     if len(series_image_path_dic[unique_image]) > 3:
#         continue 
    tmp_fetures = []    
    for item in series_image_path_dic[unique_image]:
           
        tmp = nb.load(os.path.join(base_path_prob_map, item)).get_data()
        if tmp.shape != (176, 240, 256):
            continue
        tmp_fetures.append(tmp[:, : , 128].flatten())
        
    if len(tmp_fetures) == 3:
        prob_maps.append(np.concatenate(tmp_fetures))  
        prob_maps_identifiers.append(unique_image)

prob_maps = np.array(prob_maps)
```

```python
print(prob_maps.shape)
```

```python
model_mapping_dict = {'TrioTim':0, 'Prisma_fit':1, 'Verio':2, 'Biograph_mMR':3}
```

```python
y_labels = []
for item in prob_maps_identifiers:
    mfg_model = csv_info_pd[csv_info_pd['seriesIdentifier'] == item]['Mfg Model'].item()
    y_labels.append(model_mapping_dict[mfg_model])
```

```python
len(y_labels)
```

```python
clf = RandomForestClassifier(max_depth=5)
```

```python
clf.fit(prob_maps, y_labels)
```

```python
clf.score(prob_maps, y_labels)
```

```python
clf.predict(prob_maps)
```

```python
predictions = clf.predict(prob_maps)
np_labels = np.array(y_labels)
```

```python
for i in range(4):
    correct_preds = predictions == np_labels
    precision = np.sum(np.logical_and(predictions == i, correct_preds)) / np.sum(predictions == i)
    recall = np.sum(np.logical_and(predictions == i, correct_preds)) / np.sum(np_labels == i)
    print("Label {}, precision: {}, recall: {}".format(i, precision, recall))
```
